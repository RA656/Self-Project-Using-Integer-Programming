```python
import pandas as pd
from pyomo.environ import *
```

```python
distance_matrix=pd.read_csv("final_distance.csv")
adjacency_matrix = pd.read_csv("adj_matrix_with37.csv")
```

```python
columns=distance_matrix.columns
```

```python
node_names=list(columns)[1:]
```

```python
node_names
```

```python
distance_matrix.drop(columns=["Unnamed: 0"],inplace=True)
adjacency_matrix.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
distance_matrix
```

```python
adjacency_matrix
```

```python
model=ConcreteModel()
col_indi=range(1,len(node_names)+1)
row_indi=range(1,len(node_names)+1)
```

```python
model.E=Var(row_indi,col_indi,domain=Binary)
```

```python
model.U=Var(range(1,len(node_names)+1),domain=NonNegativeIntegers,bounds=(0,len(node_names)-1))
```

```python
model.obj=Objective(expr=sum(distance_matrix.iat[i-1,j-1]*model.E[i,j] for i in row_indi for j in col_indi),sense=maximize)
```

```python
model.constraints=ConstraintList()
```

```python
for i in row_indi:
  for j in col_indi:
    model.constraints.add(model.E[i,j] <= adjacency_matrix.iat[i-1,j-1])
```

```python
model.constraints.add(model.U[1]  == 0 )
```

```python
for i in row_indi:
  model.constraints.add(sum(model.E[i,j] for j in col_indi if i!=j)<=1)
```

```python
for j in col_indi:
  model.constraints.add(sum(model.E[i,j] for i in row_indi if i!=j)== sum(model.E[j,i] for i in col_indi if i!=j))
```

```python
for i in row_indi:
  for j in col_indi:
    model.constraints.add((model.E[i,j] + model.E[j,i]) <= 1)
```

```python
for i in range(2,len(node_names)+1):
  for j in range(2,len(node_names)+1):
    if i!=j:
      model.constraints.add((model.U[i]-model.U[j]+len(node_names)*model.E[i,j])<=len(node_names)-1)
```

```python
model.pprint()
```

```python
opt_cbc=SolverFactory('cbc')
result=opt_cbc.solve(model)
```

```python
print(result)
```

```python
print('\nObjective (maximum total distance covered) = ', model.obj())
```

```python
sum_=0
for i in row_indi:
  for j in col_indi:
    if model.E[i,j].value==1:
      sum_+=1
      print(f"E[{i},{j}]={model.E[i,j].value}")
```

```python
number=[]
i=1
while len(number)!=sum_+1:
  number.append(i)
  for j in col_indi:
    if model.E[i,j].value==1:
      i=j
      break
```

```python
tour=""
var=0
for city_number in number:
  var+=1
  a="N"+str(node_names[city_number-1])
  tour+=a
  if var!=sum_+1:
    tour+="-->"
```
